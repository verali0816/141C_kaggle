## Check for missing values (also fill in)

```python
total = train.isnull().sum().sort_values(ascending=False)
percent = (train.isnull().sum()/train.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data[missing_data.Total > 0]
```

#### PoolQC 

It is pool quality which is related to PoolArea: Pool area in square feet. We can maybe match the data of these two variables and try to fill the missing values


```python
Counter(train[train.PoolQC.isnull() == True].loc[:,["PoolQC","PoolArea"]].PoolArea)
```

When PoolQC is missing, the corresponding pool area are all zero. Therefore, we can consider to fill the missing value by 0.

```python
train.PoolQC = train.PoolQC.fillna("No")
```

#### GarageType, GarageCond, GarageFinish, GarageQual, GarageYrBlt

They have the same amoubt of missing value

For the categorical variable, we could set it to be "None"
For the continuous variable, we could set it to be 0.

```python
garage_cols=['GarageType','GarageQual','GarageCond','GarageYrBlt','GarageFinish','GarageCars','GarageArea']
train[garage_cols][train['GarageType'].isnull()==True]
```

```python
garage_cate =['GarageType','GarageFinish','GarageQual','GarageCond',]
garage_cont = ['GarageYrBlt']
train[garage_cate] = train[garage_cate].fillna("None")
train[garage_cont] = train[garage_cont].fillna(0)
```

Sales price is highly correlated with some variables. Some dependent variables are also highly correlated (i.e. r > 0.8)

```python
print "The correlation between dependent variables"
correlations=train.corr()
attrs = correlations.iloc[:-1,:-1] # all except target

threshold = 0.6
important_corrs = (attrs[abs(attrs) > threshold][attrs != 1.0]) \
    .unstack().dropna().to_dict()

unique_important_corrs = pd.DataFrame(
    list(set([(tuple(sorted(key)), important_corrs[key]) \
    for key in important_corrs])), 
        columns=['Attribute Pair', 'Correlation'])

    # sorted by absolute value
unique_important_corrs = unique_important_corrs.ix[
    abs(unique_important_corrs['Correlation']).argsort()[::-1]]

unique_important_corrs
```

The above table showed that the multicollinearity is severe.

Multicollinearity increases the standard errors of the coefficients and also makes some variables statistically insignificant when they should be significant.

There are several way to deal with it
1. Remove highly correlated predictors from the model
2. Do PCA
